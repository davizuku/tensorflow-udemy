# Classification Example

```python
import pandas as pd
```

```python
diabetes = pd.read_csv('./02-TensorFlow-Basics/pima-indians-diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
cols_to_norm = ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
diabetes[cols_to_norm] = diabetes[cols_to_norm].apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
```

```python
diabetes.head()
```

```python
import tensorflow as tf
```

```python
diabetes.columns
```

```python
num_preg = tf.feature_column.numeric_column('Number_pregnant')
plasma_gluc = tf.feature_column.numeric_column('Glucose_concentration')
dias_press = tf.feature_column.numeric_column('Blood_pressure')
tricep = tf.feature_column.numeric_column('Triceps')
insulin = tf.feature_column.numeric_column('Insulin')
bmi = tf.feature_column.numeric_column('BMI')
diabetes_pedigree = tf.feature_column.numeric_column('Pedigree')
age = tf.feature_column.numeric_column('Age')
```

```python
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list(
    'Group',
    ['A', 'B', 'C', 'D']
)
```

```python
'''
assigned_group = tf.feature_column.categorical_column_with_hash_bucket(
    'Group',
    hash_bucket_size = 10
)
'''
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes['Age'].hist(bins=20)
```

```python
age_bucket = tf.feature_column.bucketized_column(
    age, 
    boundaries = [20, 30, 50, 60, 70, 80]
)
```

```python
feat_cols = [ 
    num_preg, 
    plasma_gluc, 
    dias_press, 
    tricep, 
    insulin, 
    bmi, 
    diabetes_pedigree, 
    assigned_group, 
    age_bucket
]
```

```python
# TRAIN TEST SPLIT
```

```python
x_data = diabetes.drop('Class', axis=1)
```

```python
x_data.head()
```

```python
labels = diabetes['Class']
```

```python
labels.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, 
    labels, 
    test_size=0.3, 
    random_state=101
)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(
    x=X_train, 
    y=y_train,
    batch_size=10, 
    num_epochs=1000,
    shuffle=True
)
```

```python
model = tf.estimator.LinearClassifier(
    feature_columns = feat_cols,
    n_classes = 2
)
```

```python
model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(
    x=X_test,
    y=y_test,
    batch_size=10, 
    num_epochs=1,
    shuffle=False
)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(
    x=X_test,
    batch_size=10,
    num_epochs=1,
    shuffle=False
)
```

```python
predictions = model.predict(pred_input_func)
```

```python
my_pred = list(predictions)
```

```python
my_pred
```

```python
dnn_model = tf.estimator.DNNClassifier(
    hidden_units = [10, 10, 10],
    feature_columns = feat_cols,
    n_classes = 2
)
```

```python
dnn_model.train(
    input_fn=input_func,
    steps = 1000
)
```

```python
embedded_group_col = tf.feature_column.embedding_column(
    assigned_group,
    dimension=4
)
```

```python
feat_cols = [ 
    num_preg, 
    plasma_gluc, 
    dias_press, 
    tricep, 
    insulin, 
    bmi, 
    diabetes_pedigree, 
    embedded_group_col, # needed for DNN!
    age_bucket
]
```

```python
train_input_func = tf.estimator.inputs.pandas_input_fn(
    x=X_train, 
    y=y_train, 
    batch_size=10, 
    num_epochs=1000, 
    shuffle=True
)
```

```python
dnn_model = tf.estimator.DNNClassifier(
    hidden_units = [10, 10, 10],
    feature_columns = feat_cols, 
    n_classes = 2
)
```

```python
dnn_model.train(input_fn = train_input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(
    x = X_test, 
    y = y_test, 
    batch_size = 10, 
    num_epochs = 1, 
    shuffle = False
)
```

```python
evaluation = dnn_model.evaluate(input_fn = eval_input_func, steps = 1)
```
